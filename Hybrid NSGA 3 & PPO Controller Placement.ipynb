```python
import networkx as nx
import numpy as np
import pandas as pd
from pymoo.core.problem import Problem
from pymoo.algorithms.moo.nsga3 import NSGA3
from pymoo.optimize import minimize
from pymoo.util.ref_dirs import get_reference_directions

# Load Zoo Topology dataset
graphml_file = "archive/Abilene.graphml"
G = nx.read_graphml(graphml_file)

# Extract nodes
nodes = list(G.nodes)
num_nodes = len(nodes)

# Compute Centrality Measures
degree_centrality = nx.degree_centrality(G)
betweenness_centrality = nx.betweenness_centrality(G)
closeness_centrality = nx.closeness_centrality(G)

# Constants
LINK_DELAY = 5  # ms per hop
BASE_POWER = 50  # W per controller
TRAFFIC_FACTOR = 10  # W per Gbps of traffic handled

# Generate node metrics
data = []
for node in nodes:
    # Latency: Based on Closeness Centrality
    latency = (1 / closeness_centrality[node] + 1e-6) * LINK_DELAY if closeness_centrality[node] > 0 else np.inf

    # Traffic Load: Degree Centrality scaled
    #traffic_load = degree_centrality.get(node, 0) * 10 
    traffic_load = degree_centrality[node] * 10

    # Energy Consumption: Base power + traffic impact
    #energy = BASE_POWER + (traffic_load * TRAFFIC_FACTOR)
    energy = BASE_POWER + (traffic_load * TRAFFIC_FACTOR)

    data.append([node, latency, energy, traffic_load, 
                 degree_centrality[node], betweenness_centrality[node], closeness_centrality[node]])

# Create DataFrame
df = pd.DataFrame(data, columns=["Node", "Latency (ms)", "Energy (W)", "Traffic Load",
                                 "Degree Centrality", "Betweenness Centrality", "Closeness Centrality"])

# Convert node labels to numeric indices for optimization
node_index_mapping = {node: idx for idx, node in enumerate(nodes)}
```

Step 3: Define the Optimization Problem

```python
class ControllerPlacementProblem(Problem):
    def __init__(self, num_controllers=3):
        super().__init__(n_var=num_controllers, 
                         n_obj=3,  # Latency, Energy, Load Balance
                         n_constr=0, xl=0, xu=num_nodes-1)
        self.num_controllers = num_controllers
    
    def _evaluate(self, X, out, *args, **kwargs):
        latency_vals = []
        energy_vals = []
        load_vals = []

        for solution in X:
            controllers = [nodes[int(i)] for i in solution]

            # Compute latency
            total_latency = 0
            for switch in nodes:
                min_latency = min(nx.shortest_path_length(G, source=switch, target=ctrl) * LINK_DELAY 
                                  for ctrl in controllers)
                total_latency += min_latency
            avg_latency = total_latency / len(nodes)

            # Compute energy consumption
            energy = sum(BASE_POWER + (len(list(G.neighbors(ctrl))) * TRAFFIC_FACTOR) for ctrl in controllers)



    # traffic_loads = np.random.poisson(lam=5, size=len(controllers))
    # traffic_loads = np.nan_to_num(traffic_loads)  # Replace NaN with zero
    # load_variance = -np.var(traffic_loads)


            
            # Compute load balancing
            traffic_loads = np.random.poisson(lam=5, size=len(controllers))
            load_variance = -np.var(traffic_loads)

            latency_vals.append(avg_latency)
            energy_vals.append(energy)
            load_vals.append(load_variance)

        out["F"] = np.column_stack([latency_vals, energy_vals, load_vals])
```

Step 4: Run NSGA-III Optimization

```python
# Reference directions for NSGA-III
ref_dirs = get_reference_directions("das-dennis", 3, n_partitions=12)

# Run NSGA-III
algorithm = NSGA3(ref_dirs=ref_dirs)
problem = ControllerPlacementProblem(num_controllers=3)

res = minimize(problem, algorithm, ('n_gen', 100), verbose=True)

# Get best controller placements
best_placement = res.X[np.argmin(res.F[:, 0])]  # Minimize latency
selected_controllers = [nodes[int(i)] for i in best_placement]
print("Optimized Controller Placement:", selected_controllers)
```
